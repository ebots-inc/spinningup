# Run a grid experiment on an algorithm 

```python
from spinup.utils.run_utils import ExperimentGrid
from spinup import ddpg
import tensorflow as tf

import pybullet
import pybullet_envs
```

```python
pybullet.connect(pybullet.DIRECT)
```

```python
num_runs = 1
num_cpu = 4

# env_name = "LunarLander-v2"
# env_name = 'CartPole-v0'
# env_name = 'HumanoidBulletEnv-v0'
env_name = 'LunarLanderContinuous-v2'
eg = ExperimentGrid(name='LunarLanderContinuous-v2-ddpg-bench')
eg.add('env_name', env_name, '', True)
eg.add('seed', [10*i for i in range(num_runs)])
eg.add('epochs', 100)
eg.add('steps_per_epoch', 4000)
eg.add('ac_kwargs:hidden_sizes', [(32,), (64,64)], 'hid')
eg.add('ac_kwargs:activation', [tf.tanh, tf.nn.relu], '')
eg.run(ddpg, num_cpu=num_cpu)
```

```python
pybullet.disconnect()
```
